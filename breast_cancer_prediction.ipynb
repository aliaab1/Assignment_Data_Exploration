```python
# Set Axis and drop diagnosis
x=data.drop(columns=["diagnosis"])
y=data["diagnosis"]
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.feature_selection import SelectKBest, f_classif
from sklearn.decomposition import PCA


#Feature Scaling
k=10 
scaler = StandardScaler()
x = scaler.fit_transform(x)

# Feature Selection
selector = SelectKBest(f_classif, k=k)
x = selector.fit_transform(x, y)

#Feature Extraction
pca = PCA(n_components=k)
x = pca.fit_transform(x)
x = pca.transform(x)
```

```python
# Split dataset into train and test
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,shuffle=True)
```

**Random Forest Algortihm**

```python
from sklearn.ensemble import RandomForestClassifier

#Initializing and Training Random Forest Classifier
random_forest = RandomForestClassifier()

random_forest.fit(x_train, y_train)
labels=random_forest.predict(x_test)
score=random_forest.score(x_test, y_test)

print("Score of Random Forest:", score)
```

```python

from sklearn.model_selection import GridSearchCV

#Searchin in Grid for best parameters and Accuracy
parameters={
    #"max_features": ["auto","sqrt", "log2"],
    "criterion" : ["gini", "entropy"],
    #"max_depth": [5, 10, 15],
    "min_samples_split": [2, 5, 10],
    "min_samples_leaf":[ 10, 20, 50]
}

grid_search= GridSearchCV(
    estimator= random_forest,
    param_grid= parameters,
    scoring= "accuracy",
    cv= 5
)

grid_search= grid_search.fit(x,y)

print(f"Best Parameters for Random Forest: {grid_search.best_params_} Accuracy of Random Forest Algorithm:{grid_search.best_score_} % ")
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
import matplotlib.pyplot as plt


y_prediction = random_forest.predict(x_test)

# Creating Confusion Matrix
cm = confusion_matrix(y_test, y_prediction)

sns.heatmap(cm,annot=True,cmap="Pastel1",linewidths=2,fmt="d")
plt.title("Confusion Matrix for Random Forest Algorithm")
plt.xlabel("Predicted")
plt.ylabel("True")

plt.show()
```

```python
from sklearn.model_selection import cross_val_score


clf = RandomForestClassifier(n_estimators=100, random_state=42)

# Performing Cross Validation
scores = cross_val_score(clf, x, y, cv=5)

print("Scores:", scores)
print("Mean score:", scores.mean())
```

```python
from sklearn.ensemble import RandomForestClassifier
import mlflow
from sklearn.metrics import accuracy_score, precision_score, f1_score


# running mlflow
for min_samples_leaf in [20, 50, 100]:
    with mlflow.start_run() as run:
        random_forest = RandomForestClassifier(min_samples_leaf=min_samples_leaf, random_state=42)
        random_forest.fit(x_train, y_train)
        y_pred=random_forest.predict(x_test)

        f1 = f1_score(y_test, y_pred)
        precision = precision_score(y_test, y_pred)
        accuracy=accuracy_score(y_test,y_pred)

    # Logging parameter and metrics 
        mlflow.log_param("min_samples_leaf", min_samples_leaf)
        #mlflow.log_param("max_depth", random_forest.max_depth)
        #mlflow.log_param("min_samples_split", random_forest.min_samples_split)
        mlflow.log_metric("accuracy", accuracy)
        mlflow.log_metric("f1_score", f1)
        mlflow.log_metric("precision", precision)

    # Logging model into mlfow
        mlflow.sklearn.log_model(random_forest, "random_forest_model")
        


```
